```python
%matplotlib inline
import matplotlib.pyplot as plt
import os
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import slidingwindow as sw
import cv2
```

```python
# Loading image destinations with their labels
path = 'data/CRYSTAL/'
files = [f for f in os.listdir(path)]
x, y = [], []
for i in range(0, len(files)):
    file_name = os.path.splitext(os.path.basename(files[i]))[0]
    y.append(file_name.split('_')[0])
    x.append(os.path.join(path, files[i]))
y = np.reshape(y, (-1, 1))
encoder = OneHotEncoder(handle_unknown='ignore')
encoder.fit(y)
```

```python
# Visualizing sliding window
img = cv2.imread('test.jpg')
(h, w, c) = np.shape(img)

print("Dimensions before transformation : ", w, h, c)

if h < 256:
    ratio = 256 / h
    h = 256
    w = int(w * ratio)
else:
    ratio = h / 256
    h = 256
    w = int(w / ratio)

new_img = cv2.resize(img, (w, h))
windows = sw.generate(new_img, sw.DimOrder.HeightWidthChannel, 256, 0.5)

print("Dimensions after transformation : ", w, h, c)

plt.imshow(img)
plt.title('Actual Image')
plt.show()

plt.imshow(new_img)
plt.title('Transformed image')
plt.show()

fig, axs = plt.subplots(len(windows))
fig.suptitle('Windows')
for i in range(0, len(windows)):
    axs[i].imshow(new_img[windows[i].indices()])

plt.show()
```

```python
#Preparing the dataset
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.333, random_state=44)
del x, y
```

```python
batch_number_train = 0
batch_number_test = 0

# Creating data generators feeding 8 images at a time
def train_generator(train_x, train_y):
    global batch_number_train
    start_index = (batch_number_train-1)*8
    end_index = (batch_number_train)*8-1
    batch_number_train += 1
    x, y = [], []
    
    for i in range(start_index, end_index+1):
        img = cv2.imread(train_x[i])
        # Perform all the transformations and append windows to x and y
        (h, w, c) = np.shape(img)
        if h < 256:
            ratio = 256 / h
            h = 256
            w = int(w * ratio)
        else:
            ratio = h / 256
            h = 256
            w = int(w / ratio)
        new_img = cv2.resize(img, (w, h))
        windows = sw.generate(new_img, sw.DimOrder.HeightWidthChannel, 256, 0.5)
        for j in range(0, len(windows)):
            x.append(new_img[windows[j].indices()])
            y.append(train_y[i])
            
    y = encoder.transform(y) # Make OneHot
    return x, y

def validate_generator(batch_number, test_x, test_y):
    global batch_number_test
    start_index = (batch_number_test-1)*8
    end_index = (batch_number_test)*8-1
    batch_number_test += 1
    x, y = []
    
    for i in range(start_index, end_index+1):
        img = cv2.imread(train_x[i])
        # Perform all the transformations and append windows to x and y
        (h, w, c) = np.shape(img)
        if h < 256:
            ratio = 256 / h
            h = 256
            w = int(w * ratio)
        else:
            ratio = h / 256
            h = 256
            w = int(w / ratio)
        new_img = cv2.resize(img, (w, h))
        windows = sw.generate(new_img, sw.DimOrder.HeightWidthChannel, 256, 0.5)
        for j in range(0, len(windows)):
            x.append(new_img[windows[j].indices()])
            y.append(train_y[i])
    
    y_true = encoder.transform(y)
    y_pred = model.predict(x)
    return x, y_true, y_pred
```

```python
# Test generators
temp_x, temp_y = validate_generator(train_x, train_y)
print(np.shape(temp_x), np.shape(temp_y), temp_y[0])
```
